# Naïve Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

## Base credit data - 93.80%

```python
import pickle
with open('/content/credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
X_credit_teste.shape, y_credit_teste.shape
```

```python
# treinando o algoritmo
naive_credit_data = GaussianNB()
naive_credit_data.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
# Fazendo previsoes utilizando os dados de teste
previsoes = naive_credit_data.predict(X_credit_teste)
```

```python
previsoes
```

```python
# Vamos comparar o previsto por esse resultado real da base de teste
y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Ver quanto o algoritmo acertou: 93%
accuracy_score(y_credit_teste, previsoes)
```

```python
# Matriz de confusão
confusion_matrix(y_credit_teste, previsoes)
# 428 clientes que pagam e que foram classificados como pagam
# 8 clientes que pagam e que foram classificados como nao pagam
# 23 clientes que nao pagam e foram classificados como que pagam
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
# Para facilitar a visualização
cm = ConfusionMatrix(naive_credit_data)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_teste, y_credit_teste)
```

```python
print(classification_report(y_credit_teste, previsoes))
# o algoritmo consegue identificar corretamente 98% dos clientes que pagam
# Mas o algoritmo nao foi tao bom em identificar as pessoas que nao pagam. Só consegue identificar corretamente 64% dos registros que nao pagam
```
